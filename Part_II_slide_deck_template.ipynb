## Investigation Overview


The project aim is to find out the duration when most bikes are used by which kind of customers with focus on age, time, user type and gender.  

From the exploration, it's evident that most bike users were subscribers and most bike hiring was done at 0800hrs and 1700hrs. Majority of bikers are male as compared to other genders and age distribution of most bikers is between 20-40 years.


## Dataset Overview

The Ford-GoBike dataset has 183412 records and 16 columns. 


```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
data = pd.read_csv('df_clean.csv')
data.head()
```

## Ford-GoBike-System Bike Hiring Rate

- Check the times of the day when most bikes are hired.
    * Most bikes are used in rush hours. 
    * The users may be working class and hire bikes to and from work.

```python
plt.figure(figsize=(10,6))
data.groupby('Hour')['start_time'].count().plot(kind='line')
plt.xticks([i for i in range(0,25)])
plt.xlabel('Hours')              
plt.ylabel('Number of Bikes Hired')
plt.title('Hourly Rate Bike Hires')
plt.show();
```

## Gender Distribution of Customers

- Cheking on the gender of most users.
    * most biker users are male subscribers.

```python
plt.figure(figsize = (10,7))
sb.countplot(data=data, x= 'user_type', hue ='member_gender')
plt.title('Gender - user type Distribution')
plt.xticks(rotation=15);
```

## Age and Duration correlation of users types



```python
data['duration_hours'] = data['duration_sec']/3600
plt.figure(figsize=(15,7))
sb.scatterplot(x= data['customer_age'], y=data['duration_hours'] , hue=data['user_type'])
plt.title('Distribution of User Type Considering Age and Duration Spent in Hours')
plt.xticks([i for i in range(0,150,20)])
plt.show()
```

Most users are between 20 - 40 years and majorly spend less than 5 hours biking.
